```python
df=pd.DataFrame(data)

Positive=df[df['Years gained / lost']>0]
Positive
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(Positive)
plt.figure(figsize=(12, 6))
plt.bar(df['Factor'], df['Years gained / lost'], color='red')
plt.xticks(rotation=90)
plt.xlabel('Factor')
plt.ylabel('Years gained / lost')
plt.title('Distribution of Years gained / lost by Factor')
plt.tight_layout()
```

```python
negative = data[data['Years gained / lost'] < 0]
negative
```

```python
df = pd.DataFrame(negative)
plt.figure(figsize=(12, 6))
negative_values = df[df['Years gained / lost'] < 0]
plt.bar(negative_values['Factor'], negative_values['Years gained / lost'], color='yellow', label='Negative')
plt.xticks(rotation=90)
plt.xlabel('Factor')
plt.ylabel('Years gained / lost')
plt.title('Distribution of Years gained / lost by Factor')
plt.legend()
plt.tight_layout()
```
